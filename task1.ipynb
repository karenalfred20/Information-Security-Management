```python
import itertools
import time
import tkinter as tk
from tkinter import filedialog, messagebox

# Hardcoded correct username and password
CORRECT_USERNAME = "admin"
CORRECT_PASSWORD = "apple"

# Load passwords from dictionary file
def load_dictionary(file_path):
    try:
        with open(file_path, "r", encoding="utf-8", errors="ignore") as file:
            return [line.strip() for line in file.readlines()]
    except FileNotFoundError:
        messagebox.showerror("Error", "Dictionary file not found! Make sure the path is correct.")
        return []

# Dictionary Attack
def dictionary_attack(username, dictionary_file, output_text):
    output_text.insert(tk.END, f"\nüîç Attempting Dictionary Attack for user: {username}\n")
    
    password_list = load_dictionary(dictionary_file)
    if not password_list:
        output_text.insert(tk.END, "\n‚ùå Dictionary file is empty or not found. Skipping attack...\n", "red")
        return False

    for password in password_list:
        output_text.insert(tk.END, f"üîë Trying password: {password}\n")
        output_text.update()
        time.sleep(0.01)  # Simulating delay like a real attack
        
        if username == CORRECT_USERNAME and password == CORRECT_PASSWORD:
            output_text.insert(tk.END, f"\n‚úÖ SUCCESS! The correct password is: {password}\n", "green")
            return True
    
    output_text.insert(tk.END, "\n‚ùå Dictionary Attack Failed! Moving to Brute Force Attack...\n", "red")
    return False

# Brute Force Attack
def brute_force_attack(output_text):
    output_text.insert(tk.END, "\nüöÄ Starting Brute Force Attack...\n")
    output_text.update()

    characters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
    attempts = 0
    start_time = time.time()

    for combination in itertools.product(characters, repeat=5):
        password = "".join(combination)
        attempts += 1

        if attempts % 100000 == 0:
            output_text.insert(tk.END, f"‚åõ Attempts so far: {attempts}...\n")
            output_text.update()

        if password == CORRECT_PASSWORD:
            end_time = time.time()
            output_text.insert(tk.END, f"\n‚úÖ Password found: {password}\n", "green")
            output_text.insert(tk.END, f"‚è≥ Total attempts: {attempts}\n")
            output_text.insert(tk.END, f"‚è±Ô∏è Time taken: {round(end_time - start_time, 2)} seconds\n")
            return

# Start attack process
def start_attack():
    username = username_entry.get()
    dictionary_file = file_path_entry.get()
    
    if not username or not dictionary_file:
        messagebox.showwarning("Input Error", "Please enter a username and select a dictionary file!")
        return
    
    output_text.delete("1.0", tk.END)  # Clear previous output
    output_text.tag_config("red", foreground="red")
    output_text.tag_config("green", foreground="green")
    
    if not dictionary_attack(username, dictionary_file, output_text):
        brute_force_attack(output_text)

# Select file function
def select_file():
    file_path = filedialog.askopenfilename(title="Select Dictionary File", filetypes=[("Text Files", "*.txt")])
    file_path_entry.delete(0, tk.END)
    file_path_entry.insert(0, file_path)

# Create GUI
root = tk.Tk()
root.title("Password Cracker")
root.geometry("600x550")
root.resizable(False, False)

tk.Label(root, text="üë§ Enter Username:", font=("Arial", 12)).pack(pady=5)
username_entry = tk.Entry(root, font=("Arial", 12))
username_entry.pack(pady=5)

tk.Label(root, text="üìÇ Select Dictionary File:", font=("Arial", 12)).pack(pady=5)
file_frame = tk.Frame(root)
file_frame.pack()
file_path_entry = tk.Entry(file_frame, font=("Arial", 12), width=40)
file_path_entry.pack(side=tk.LEFT, padx=5)
tk.Button(file_frame, text="Browse", font=("Arial", 10), command=select_file).pack(side=tk.RIGHT)

tk.Button(root, text="üöÄ Start Attack", font=("Arial", 14, "bold"), bg="red", fg="white", command=start_attack).pack(pady=10)

output_text = tk.Text(root, font=("Arial", 10), height=15, width=70)
output_text.pack(pady=5)

root.mainloop()
```
